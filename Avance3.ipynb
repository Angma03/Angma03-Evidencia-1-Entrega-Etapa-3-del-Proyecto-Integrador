## Descripción: 

Según datos de la OCDE (Organización para la Cooperación y el Desarrollo Económico) extraídos del programa para la evaluación internacional de alumnos, en la prueba que se realizó en el año 2018, los estudiantes mexicanos obtuvieron un puntaje por debajo del promedio de los países que pertenecen a la OCDE en las áreas de lectura, matemáticas y ciencias. 

Los resultados de la evaluación muestran un escaso avance con respecto a los años anteriores, aunque se muestre un avance sigue sin ser suficiente y muestra que todavía falta mucho por trabajo por hacer para mejorar el nivel y la calidad de educación que hay en México.

Ante esta situación se nos pregunta, ¿cómo podemos implementar los programas computacionales para apoyar a los estudiantes de nivel básico y medio? 

Mediante el uso del lenguaje de programación conocido como Python crearemos un programa que ayude a los alumnos a estudiar y prepararse de una mejor manera para estas evaluaciones, de esta manera lograr mejores resultados en las pruebas y mostrar como las ciencias computacionales pueden ayudarnos a mejorar. 

Para poder crear este programa usaremos los conocimientos que hemos ido aprendiendo durante el curso de pensamiento computacional para ingeniería, donde tendremos que usar funciones como lo son los ciclos while, valores booleanos, y muchos más que nos permitirán desarrollarlo de manera efectiva y funcional.    

El programa por desarrollar presentará un menú el cuál contará con 5 opciones, que le permitirá al usuario prepararse. Cuando el programa inicia se desplegará el menú, y se le preguntará al usuario función querrá usar, teniendo así la opción de seleccionar las siguientes:
1.	Dar de alta preguntas con sus respuestas de la prueba PISA relacionada a los ámbitos de lectura, matemáticas y ciencias.
2.	Actualizar las preguntas de la prueba PISA
3.	Estudiar las preguntas de alguna área en especifico (lectura, matemáticas o ciencias).
4.	Presentar un quiz simulador de una prueba PISA donde se pone a prueba sus conocimientos adquiridos, que servirá de práctica para la prueba real.
5.	Ingresar al reporte de calificaciones el cuál mostrará los resultados del quiz presentado en la función anterior
6.	Y finalmente tendrá la opción de salir del programa.

Si bien esta propuesta no demuestra una solución absoluta a la problemática, si aporta mucho al camino de mejora continua en la educación, presenta una herramienta nueva para poder estudiar, y demuestra lo versátil que pueden llegar a ser los programas y las ciencias computacionales.

# Algoritmos

### Alta de preguntas:
Función la cual te permitirá guardar los datos de la pregunta que quieras dar de alta, por cada pregunta que des de alta, deberás ingresar el id, enfoque, pregunta, opciones de respuesta y la respuesta correcta, todo esto deberá almacenarse en una lista, y cada pregunta que des de alta se almacenará en listas distintas que a su vez se almacenarán en una lista principal.

### Actualizar preguntas:
Una vez inicializada la función podrás observar todas las preguntas que fueron dadas de alta, posteriormente se te preguntará el id de la pregunta que quieres actualizar, una vez que ingreses el id, el programa validará la existencia del mismo y en caso de que si exista se te desplegará toda la información de esa pregunta, y se te preguntará elemento por elemento si quieres editarlo, si algun elemento no deseas editar solamente deberás darle enter y continuar, una vez que termines se te mostrará la pregunta actualizada y se te volverá a preguntar si estás seguro que quieres actualizar los datos de la pregunta. Si la respuesta es Si, la función remplazará los datos anteriores con los nuevos en la lista principal.

### Estudiar preguntas área:
Al iniciar la función se te preguntará que área quieres estudiar y cuantas preguntas quieres responder en el quiz de práctica, posteriormente la función analizará el área escogida y de la lista principal extraerá todas las preguntas relacionadas al área para después escoger la cantidad de preguntas que se te serán preguntas según tu petición y se te desplegarán en formato de quiz el cual podrás responder y usar de práctica, para ver si tuviste bien o mal la respuesta.

### Presentar un quiz:
Al inicializar la función se te irán desplegando de la lista principal conforme respondas, 10 preguntas con opción multiple y tu deberás teclear la respuesta.  Las preguntas serán de las 3 áreas de estudios previamente establecidas, mientrás respondes contarás con un límite de tiempo de 20 minutos para poder terminar el quiz.

### Reporte de calificaciones:
Una vez iniciada la función, se te desplagará una tabla con diferente información. Primero se te mostrará la cantidad de usuarios que desplegaron la función de presentar un quiz, después te mostrará el promedio de calificaciones obtenidas en el examen, y por último podrás ver el porcentaje total de las preguntas contestadas correctas e incorrectamente.

### Salir:
Al iniciar la opción salir, automáticamente el programa se detendrá y se cerrará siendo así que ya terminaste de hacer uso del mismo y no podrás usar las funciones al menos que vuelvas a iniciar el programa.

```python
from random import choice

contador = 0
contador_correctas = 0
contador_incorrectas = 0

global nombre_archivo
nombre_archivo = "preguntas_guardadas.txt"

preguntas = []

def guardar_archivo(nombre):

  with open(nombre, "w") as archivo:

    for pregunta in preguntas:

      pregunta[0] = str(pregunta[0])

      texto = ",".join(pregunta)
      archivo.write(texto + "\n")


guardar_archivo(nombre_archivo)

def leer_archivo(nombre):

  with open(nombre, "r") as archivo:

    contenido = archivo.readlines()

    for elemento in contenido:
      linea = elemento.split(",")
      linea[7] = linea[7][:-1]
      preguntas.append(linea)

    return preguntas

preguntas = leer_archivo(nombre_archivo)


def registrar_preguntas():

  
  global nombre_archivo
  global preguntas

  id_pregunta = int(input("Ingresa el id de la pregunta: "))
  enfoque = input("¿Cuál es el enfoque de tu pregunta? (Matemáticas, Lectura, Ciencias): ")
  pregunta = input("Inserta tu pregunta: ")
  respuestas = ""

  for num in range(4):
    respuesta = input("Ingresa la respuesta " + str(num + 1) + ": ")

    if num == 3:
      respuestas = respuestas + respuesta 
    else:
      respuestas = respuestas + respuesta + "," #string de respuestas (todas)

  respuestas = respuestas.split(",") # hacer una lista desde un string, cada elemento es separado por coma

  respuesta1, respuesta2, respuesta3, respuesta4 = respuestas #Asignación multiple de variables de una lista (elemento por elemento)
  
  correcta = input("Ingresa la respuesta correcta:")

  pregunta = [id_pregunta, enfoque, pregunta, respuesta1, respuesta2, respuesta3, respuesta4, correcta]
  
  """print("Pregunta registrada exitosamente")
  print("Enfoque de la pregunta:", enfoque)
  print("Pregunta:", pregunta)
  print("Respuestas", respuestas)
  print("La respuesta correcta es", correcta)"""

  preguntas.append(pregunta)

  guardar_archivo(nombre_archivo)
  
def actualizar_preguntas():

  global preguntas
  global nombre_archivo

  print("Bienvenido a la opción para actualizar las preguntas, por favor sigue las instrucciones")

  for pregunta in preguntas:
    
    id_pregunta, enfoque, pregunta, respuesta1, respuesta2, respuesta3, respuesta4, correcta = pregunta # esto (pregunta) es una lista, hacemos asignación múltiple

    print(f"ID de la pregunta: {id_pregunta} \nEnfoque de la pregunta: {enfoque} \nPregunta: {pregunta}")
    print(f"a){respuesta1:<15} b){respuesta2:<15} c){respuesta3:<15} d){respuesta4:<15} \n")

  id_cambio = int(input("Ingresa el id de la pregunta que quieres cambiar: "))
  
  if id_cambio >= 1 and id_cambio <= len(preguntas): # verifica que el id de la pregunta exista
    print(preguntas[id_cambio - 1]) #Si sí existe entonces imprime la información de ese id
    cambio_pregunta = preguntas[id_cambio-1] #genero una variable con la lista seleccionada
    
    for i in range(1, len(cambio_pregunta)): #genero un ciclo for donde me va a ir mostrando cada elemento de la lista a excepción del id 
      print(cambio_pregunta[i]) 

      nuevo_dato = input('Inserta el nuevo dato ') # Me pedirá insertar el nuevo dato por el cual cambiará, si no quiero cambiar el dato de ese elemento solamente doy "enter" 
      if nuevo_dato != '':    #Si mi respuesta es diferente a enter (espacio en blanco), el nuevo dato sustituirá al elemento de la lista que está en la variable
        cambio_pregunta[i] = nuevo_dato
    print(cambio_pregunta) #Me muestra como quedaría la nueva pregunta
    mod = input("Quieres modificar los datos? (S/N) ") #Me pregunta si quiero realizar los cambios
    if mod == "S": 
      preguntas[id_cambio-1] = cambio_pregunta #Si la respuesta es sí, entonces sustituyo los datos que están en mi lista por la variable generada
      print(preguntas[id_cambio-1])
      print("Cambios realizados con éxito")
    else: 
      print("No se realizó ningún cambio")
  else:
    print("El id ingresado no existe")

  guardar_archivo(nombre_archivo)

def estudiar_area():

  global preguntas 

  preguntas_area = []

  enfoque = input("¿Qué area deseas estudiar? (Matemáticas, Lectura, Ciencias) ")

  for pregunta in preguntas:
    if pregunta[1] == enfoque:
      preguntas_area.append(pregunta)

  print(f"Hay {len(preguntas_area)} preguntas de esta área")
  cant_quiz = int(input("¿Cuántas preguntas deseas estudiar? "))
  
  for num in range(cant_quiz):

    pregunta = preguntas_area[num]

    print(f"ID de la pregunta: {pregunta[0]} \nEnfoque de la pregunta: {pregunta[1]} \nPregunta: {pregunta[2]}")
    print(f"a){pregunta[3]:<15} b){pregunta[4]:<15} c){pregunta[5]:<15} d){pregunta[6]:<15} \nRespuesta correcta: {pregunta[7]} \n")

def presentar_quiz():

  global preguntas
  global contador
  global contador_correctas
  global contador_incorrectas

  contador = contador + 1

  print(f"Bienvenido al quiz, tienes 20 minutos para completarlo \n")

  lista_quiz = [] # crea una lista para almacenar las preguntas sin repetir.
  contador_rondas = 0 # almacena las veces que se agrega una pregunta a la lista.

  lista_random = choice(preguntas) # genera una pregunta al azar de la lista de preguntas
# El código para randomizar el quiz fue obtenido de stack overflow
# se repite hasta que el tamaño de la lista no llegue a 10. 
  while contador_rondas < 10:
    lista_random = choice(preguntas) # genera una pregunta al azar de la lista de preguntas
    if lista_random in lista_quiz: # verifica si la pregunta ya se encuentra en la lista
        pass
    else:
        lista_quiz.append(lista_random) # agrega la pregunta a la lista
        contador_rondas = contador_rondas + 1 # suma a cada vuelta

  for num in range(10):

    pregunta = lista_quiz[num]

    print(f"ID de la pregunta: {pregunta[0]} \nEnfoque de la pregunta: {pregunta[1]} \nPregunta: {pregunta[2]}")
    print(f"a){pregunta[3]:<15} b){pregunta[4]:<15} c){pregunta[5]:<15} d){pregunta[6]:<15} \n")

    respuesta = input("Ingresa tu respuesta: ")

    if respuesta == pregunta[7]:
      print(f"Respuesta correcta \n")
      contador_correctas = contador_correctas + 1
    
    else:
      print(f"Respuesta incorrecta, la respuesta correcta es {pregunta[7]} \n")
      contador_incorrectas = contador_incorrectas + 1

  print("Terminaste el Quiz, para conocer el resultado accede al reporte de calificación en el menú.")


def reportar_calificaciones():
  calificacion = ((contador_correctas * 100) / 10) / contador
  preguntas_correctas  = ((contador_correctas * 100) / 10) / contador
  preguntas_incorrectas = ((contador_incorrectas * 100) / 10) / contador
  print(f"Total de usuarios: {contador:>13} \nPromedio de calificaciones: {calificacion:>6} \nPreguntas Correctas: {preguntas_correctas:>13}% \nPreguntas incorrectas: {preguntas_incorrectas:>11}% ")




def menu():

  print(
    """Menu de opciones: 
     1. Alta de preguntas de prueba PISA 
     2. Actualizar preguntas de la prueba PISA 
     3. Estudiar preguntas de cierta área 
     4. Presentar un quiz 
     5. Reporte de calificaciones 
     6. Salir"""
  )

  opcion = int(input("Selecciona una opción del menú: "))

  return opcion


def main():

  while True:
    
    opcion = menu()

    if opcion < 1 or opcion > 6:
      print("Opción incorrecta")
      continue

    elif opcion == 1:
      registrar_preguntas()

    elif opcion == 2:
      actualizar_preguntas()

    elif opcion == 3:
      estudiar_area()

    elif opcion == 4:

      presentar_quiz()

    elif opcion == 5:
      reportar_calificaciones()

    elif opcion == 6:
      print("Programa terminado")
      break

main()
```
